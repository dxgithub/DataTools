```python
import xml.etree.ElementTree as ET
import os
import sys
import numpy as np
f=open("E:\Data\FaceG\IR_labelled\IR_Training_label.txt","w")
f.close()
f=open("E:\Data\FaceG\IR_labelled\IR_Training_label.txt","a")
L=os.listdir("E:\Data\FaceG\IR_labelled\Training")
for file in L:
    a=file.split(".")
    #print(a[-1])
    if (a[-1]=="jpg"):
        xmlFilePath = os.path.abspath("E:/Data/FaceG/IR_labelled/Training/"+a[0]+".xml")
        print("E:/Data/FaceG/IR_labelled/Training/"+a[0]+".xml")
        try:
            tree = ET.parse(xmlFilePath)
            # 
            root = tree.getroot()
        except Exception as e:  #sys.exit()
            print ("parse test.xml fail!")
            sys.exit()    
        BoxList=root.findall("object")
        bboxes=np.zeros((len(BoxList),4),dtype=np.int16)
        for i in range(len(BoxList)):
            bboxes[i,0]=int(BoxList[i][4][0].text)  
            bboxes[i,1]=int(BoxList[i][4][1].text)
            bboxes[i,2]=int(BoxList[i][4][2].text)
            bboxes[i,3]=int(BoxList[i][4][3].text)
        bboxes=bboxes.reshape(-1)
        temp=bboxes.tolist()
        labels=[]
        labels.append(''.join((a[0]+".jpg")))
        for pos in temp:
            labels.append(pos)
        f.write(str(labels).strip("[]"))
        f.write("\n")
f.close()       
```
